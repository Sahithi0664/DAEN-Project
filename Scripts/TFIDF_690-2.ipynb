```python
def remove_punctuation(text):
    no_punct=[words for words in text if words not in string.punctuation]
    words_wo_punct=''.join(no_punct)
    return words_wo_punct
mydata['new_summary']=mydata['summary'].apply(lambda x: remove_punctuation(x))
mydata.head()
```

```python
def tokenize(text):
    split=re.split("\W+",text) 
    return split
mydata['new_summary_one']=mydata['new_summary'].apply(lambda x: tokenize(x.lower()))
mydata.head()
```

```python
stopword = nltk.corpus.stopwords.words('english')
print(stopword[:15])
```

```python
def remove_stopwords(text):
    text=[word for word in text if word not in stopword]
    return text
mydata['new_summary_two'] = mydata['new_summary_one'].apply(lambda x: remove_stopwords(x))
mydata.head()
```
